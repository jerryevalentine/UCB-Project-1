```python
import pandas as pd
import numpy as np
import datetime

car_model_attributes = "data/Car Dataset 1945-2020.csv"
df_car_model_attributes = pd.read_csv(car_model_attributes)

print(datetime.datetime.now())
```

```python
# Count missing values in each column
missing_values = df_car_model_attributes.isnull().sum()
data_types = df_car_model_attributes.dtypes
unique = df_car_model_attributes.nunique()
non_null_counts = df_car_model_attributes.count()

null_counts = df_car_model_attributes.isnull().sum()
total_rows = len(df_car_model_attributes)
null_percentage = (null_counts / total_rows) * 100

dataframe_metadata = pd.DataFrame({'Non-null Counts': non_null_counts, 'missing_values' :missing_values, 'null_percentage': null_percentage ,'Data Types': data_types, 'unique': unique})
dataframe_metadata
```

```python
import pandas as pd

# Assuming df is your DataFrame

# Convert specific columns to float
columns_to_convert = ['Year_from', 'Year_to']
df_car_model_attributes[columns_to_convert] = df_car_model_attributes[columns_to_convert].astype(float)

# Drop rows with any missing values
df_car_model_attributes.dropna(subset=columns_to_convert, inplace=True)

# Optionally, convert the columns back to integer
df_car_model_attributes[columns_to_convert] = df_car_model_attributes[columns_to_convert].astype(int)
```
