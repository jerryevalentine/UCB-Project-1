```python
import pandas as pd

str_path_file = "data/car_prices.csv"

# Read the CSV file into a DataFrame
df_car_prices = pd.read_csv(str_path_file)
df_car_prices.head()
```

```python
import pandas as pd

# Assuming df_car_prices is your DataFrame containing 'year' and 'saledate'

# Convert 'saledate' to datetime object, ignoring errors
df_car_prices['saledate'] = pd.to_datetime(df_car_prices['saledate'], errors='coerce')

# Drop rows with invalid datetime values (NaT)
df_car_prices.dropna(subset=['saledate'], inplace=True)

# Extract year from 'saledate'
df_car_prices['saledate_year'] = df_car_prices['saledate'].dt.year

# Calculate age as the difference between 'year' and 'saledate_year'
df_car_prices['age'] = df_car_prices['saledate_year'] - df_car_prices['year']

# Drop 'saledate_year' column if not needed
df_car_prices.drop(columns=['saledate_year'], inplace=True)
```
